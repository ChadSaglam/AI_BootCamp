```python
page=requests.get('https://www.karar.com/yazarlar/zafer-acar')
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
```

```python
links=soup.find_all('article',{'class': 'item box-shadow'})
```

```python
links
```

```python
paragrafs=[]
headers=[]

for link in links:
    link1='https://www.karar.com'+link.a.get('href')
    header=link.h3.text
    
    page2=requests.get(link1)
    soup2=BeautifulSoup(page2.content,'html.parser')

    article=soup2.select('article.article-content')[0].text
    paragrafs.append(article)
    headers.append(header)
    
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=['Title', 'Article'])
```

```python
df['Title'] = headers
df['Article'] = paragrafs
```

```python
df.head()
```

```python
df['Article']=df['Article'].str.replace('\n','')
```

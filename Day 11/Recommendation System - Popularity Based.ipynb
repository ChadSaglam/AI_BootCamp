```python
ratings=pd.merge(movies,ratings)
```

```python
ratings
```

```python
ratings.title.value_counts()
```

```python
ratings.title.nunique()
```

```python
grouped=ratings.groupby("title").agg({"rating":[np.size,np.sum,np.mean]})
```

```python
grouped.head()
```

```python
populer=grouped.sort_values(("rating","mean"),ascending=False) #büyükten küçüğe sırala
```

```python
populer.head()
```

```python
grouped_sum=grouped["rating"]["sum"].sum()
```

```python
grouped_sum
```

```python
populer["percentage"]=populer["rating","sum"].div(grouped_sum)*100
```

```python
populer.head()
```

```python
popular_movies=populer.sort_values(("percentage"),ascending=False)
```

```python
popular_movies
```

```python
popular_movies['Rank']=popular_movies['percentage'].rank(ascending=False)
```

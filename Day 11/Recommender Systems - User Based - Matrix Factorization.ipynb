# Matrix Factorization

```python
# Predict which movie the user will like based on their comments
```

```python
movie_ratings=ratings.pivot_table(index=["user_id"],columns=["title"],values="rating")
```

```python
movie_ratings
```

```python
starwars=movie_ratings["Star Wars (1977)"]
```

```python
starwars.head()
```

```python
starwars.value_counts()
```

```python
movie_ratings[["101 Dalmatians (1996)","Star Wars (1977)"]].corr() # bir filmin diğer filme benzerliği %21 
```

```python
movie_ratings.corrwith(starwars).sort_values(ascending=False) #hangi filmlerle korelasyonu var onu görebiliyoruz
```

```python
similar_movies=movie_ratings.corrwith(starwars)
```

```python
#similar_movies=similar_movies.dropna() # Delete empty datas
```

```python
similar_movies.sort_values(ascending=False)
```

```python
movie_stats=ratings.groupby("title").agg({"rating":[np.size,np.sum,np.mean]})
```

```python
movie_stats
```

```python
popular_movies=movie_stats[movie_stats["rating"]["size"] > 100]
```

```python
popular_movies.sort_values(('rating', 'mean'), ascending=False)
```

```python
similar_movies_df = pd.DataFrame(similar_movies, columns=['Similarity'])
```

```python
similar_movies_df
```

```python
popular_movies.columns=popular_movies.columns.get_level_values(1)

df=popular_movies.join(similar_movies_df)
```

```python
df
```

```python
df.sort_values(['Similarity'], ascending=False)
```

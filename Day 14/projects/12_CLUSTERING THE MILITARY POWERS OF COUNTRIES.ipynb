```python
df=pd.read_excel("./Day 14 Students files/World military power.xlsx", skiprows=1)
```

```python
# Ülke isimlerinin tekrari olan alanlardan kurtulalim
df1 = df.iloc[:, 0:2] # Ilk 2 sütunu alalim
df2 = df.filter(like='value') # Sütun adinda value olanlar alalim
df=pd.concat([df1, df2], axis=1) # Bu ikisini birlestirelim
df.head()
```

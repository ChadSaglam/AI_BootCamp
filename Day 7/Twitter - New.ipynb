```python
df["text"]=df['text'].str.lower()
df["text"]=df['text'].str.replace('[^\w\s]','') #noktalama isaretklerini kaldir
df["text"]=df['text'].str.replace('\d+','') #Rakamlari kaldir
df["text"]=df['text'].str.replace('\n','').replace('\r','') #satirlari ve enter islemlerini kaldir
```

```python
from textblob import TextBlob

def detect_sentiment(text):
    return TextBlob(text).sentiment.polarity
```

```python
df['sentiment']=df['text'].apply(detect_sentiment())
```

```python
df.sentiment.value_counts()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#0: notr, 1: positive, 2: negative

def sentiment(score):
    if score>0.2:
        return 1
    elif score > -0.2:
        return 2
    else:
        return 0
```

```python
df['class']=df['sentiment'].apply(sentiment)
```

```python
sns.countplot(x=df['class'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
x=df["Text"]
y=df["Label"]
```

```python
vect=CountVectorizer(stop_words="english",ngram_range=(1,2))
```

```python
x=vect.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
```

```python
d=DecisionTreeClassifier()
```

```python
model=d.fit(x_train,y_train)
```

```python
guess=model.predict(x_test)
```

```python
accuracy_score(y_test, guess)
```

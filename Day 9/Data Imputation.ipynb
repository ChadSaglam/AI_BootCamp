```python
import pandas as pd
```

```python
df=pd.read_csv('./Day 9 Students files/ExampleData.csv')
```

```python
df
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
#Fills empty datas with Linear regression
lr=LinearRegression()
imp=IterativeImputer(estimator=lr)
imputed_data=imp.fit_transform(df)
```

```python
pd.DataFrame(imputed_data)
```

```python
#Fills empty datas with the nearest neighbor
from sklearn.impute import KNNImputer
```

```python
knn=KNNImputer()
```

```python
imputed_data2=knn.fit_transform(df)
```

```python
pd.DataFrame(imputed_data2, columns=[df.columns])
```

```python
#Fills empty datas with the mean
from sklearn.impute import SimpleImputer
```

```python
imp=SimpleImputer(strategy='mean') # median, # mode
```

```python
imputed_data3=imp.fit_transform(df)
```

```python
pd.DataFrame(imputed_data3)
```
